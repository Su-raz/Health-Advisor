# train test split




```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = dataset.drop("prognosis" , axis =1)
y = dataset['prognosis']
# label encoding to output y
le = LabelEncoder() # object creation
le.fit(y)
Y = le.transform(y)
```

```python
print(Y)
```

```python
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size = 0.3 ,random_state=20)
```

```python
x_train.shape , x_test.shape ,y_train.shape, y_test.shape
```

# Training different models

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score , confusion_matrix
import numpy as np
# create a dictonary to store models
models ={
    "SVC":SVC(kernel = 'linear'),
    "RandomForest":RandomForestClassifier(n_estimators = 100 , random_state=42),
    "GradientBosting":GradientBoostingClassifier(n_estimators = 100 , random_state =42),
    "KNeighbor":KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB":MultinomialNB()
}

for model_name, model in  models.items():
  # print(model_name , ":",model)
  model.fit(x_train , y_train) # train model
  #test model
  predictions = model.predict(x_test)

  #calculate accuracy
  accuracy = accuracy_score(y_test, predictions)
  #calculate confusion matrix
  cm = confusion_matrix(y_test , predictions)

  print(f"{model_name} accuracy:{accuracy}")
  print(f"{model_name} Confusion Matrix:")
  print(np.array2string(cm, separator = ','))

```

#single prediction

```python
svc = SVC(kernel ='linear')
svc.fit(x_train , y_train)
ypred = svc.predict(x_test)
accuracy_score(y_test , ypred)

```
